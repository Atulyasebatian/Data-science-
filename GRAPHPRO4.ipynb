```python


import matplotlib.pyplot as p
import numpy as np

x=np.array([1,6])
y=np.array([2,5])
p.subplot(2,3,1)
p.plot(x,y,'o')
p.title("Title1")
p.xlabel("Blue color")
p.ylabel("Blue Color")

x1=np.array([1,2,3,4,5])
y1=np.array([1,4,2,3,4])
x2=np.array([1,2,3,4,5])
y2=np.array([2,2,3,4,5])
p.subplot(2,3,2)
p.plot(x1,y1,linestyle='dashed')
p.plot(x2,y2,marker="*",linewidth=2.5)

x=np.array([1,2,3,4])
y=np.array([1,4,2,3])
p.subplot(2,3,3)
p.plot(x,y,marker="o",mfc='yellow',mec='black')

x=np.array([1,2,3,4])
y=np.array([1,2,3,4])
p.subplot(2,3,4)
p.plot(x,y)
p.title("Title 2")
p.xticks(np.arange(0,5,1))
p.yticks(np.arange(0,5,1))
p.grid()

x1=np.array([1,2,3,4,5,6,7,8,4,3])
y1=np.array([1,4,2,3,4,3,6,4,7,8])
colors=np.array(["red","yellow","blue","green","magenta","hotpink","lime","cyan","orange","brown"])
sizes=np.array([100,50,60,50,100,30,250,30,60,230])
p.subplot(2,3,5)
p.scatter(x1,y1,c=colors,s=sizes,alpha=0.5)

x=np.array(["A","B","C"])
y=np.array([6,3,8])
p.subplot(2,3,6)
p.bar(x,y,width=0.1,color='red')

p.show()

```

```python
import matplotlib.pyplot as p
import numpy as np

x=np.array([35,25,20,20])
ex=[0.2, 0, 0, 0]
c=["red","magenta","hotpink","b"]
lab=["A","B","C","D"]


p.pie(x,startangle=90,explode=ex,shadow=True,colors=c,labels=lab)
p.legend()
p.show()
```
